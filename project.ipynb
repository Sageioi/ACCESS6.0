```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv(r"C:\Users\USER\Downloads\Warehouse_and_Retail_Sales.csv")
```

```python
df.info()
df.head()
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
df['TOTAL_SALES'] = df['RETAIL SALES'] + df['WAREHOUSE SALES']
```

```python
df['QUARTER'] = df['MONTH'].apply(lambda x: (x-1)//3 + 1)
```

```python
item_type_avg = df.groupby('ITEM TYPE')['TOTAL_SALES'].transform('mean')
df['AVG_SALES_BY_TYPE'] = item_type_avg
```

```python
df[['YEAR', 'MONTH', 'ITEM TYPE', 'TOTAL_SALES', 'QUARTER', 'AVG_SALES_BY_TYPE']].head()
```

```python
df.to_csv('Engineered_Warehouse_and_Retail_Sales.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Initial Data Inspection
df = pd.read_csv(r"C:\Users\USER\Downloads\housing.csv")
print("Shape:", df.shape)
print("\nColumns and Data Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# 2. Data Cleaning
# Handling missing values in 'total_bedrooms'
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
# Checking for duplicates
print("\nDuplicates found:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# 3. Exploratory Data Analysis (EDA)
# Visualization 1: Distribution of Median House Value
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=50, kde=True, color='blue')
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.savefig('median_house_value_dist.png')
plt.close()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df, alpha=0.3, color='green')
plt.title('Median Income vs Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.savefig('income_vs_value.png')
plt.close()
```

```python
# 4. Feature Engineering
df['rooms_per_household'] = df['total_rooms'] / df['households']
df['bedrooms_per_room'] = df['total_bedrooms'] / df['total_rooms']
df['population_per_household'] = df['population'] / df['households']
```

```python
print("\nNew Features Created:")
print(df[['rooms_per_household', 'bedrooms_per_room', 'population_per_household']].head())
```

```python
# Save the cleaned and engineered dataset
df.to_csv('cleaned_housing.csv', index=False)
```
