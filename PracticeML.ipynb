```python
def analyze_transactions(transactions):
    number_of_transactions = len(transactions)

    if number_of_transactions == 0:
        print()
        return

    total_sum = sum(transactions)

    average_transaction = total_sum / number_of_transactions

 
    print(f"{total_sum:,.2f}")          
    print(f"{average_transaction:,.2f}")
    print(f"{number_of_transactions}")  


transaction_list_1 = [105.50, 45.00, 230.75, 12.99, 500.00]
print("--- Output for List 1 ---")
analyze_transactions(transaction_list_1)
print("-" * 25)

transaction_list_2 = []
print("\n--- Output for List 2 (Empty) ---")
analyze_transactions(transaction_list_2)
print("-" * 25)


transaction_list_3 = [50, 100, 25, 75]
print("\n--- Output for List 3 (Whole Numbers) ---")
analyze_transactions(transaction_list_3)
print("-" * 25)
```
